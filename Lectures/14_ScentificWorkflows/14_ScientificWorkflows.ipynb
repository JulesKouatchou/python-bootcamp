# System Identification

When modeling brain data, it's useful to think of it in a system identification framework:

Your job is to create a black box which, when given the stimulus, outputs a signal that is maximally correlated with the signal recorded from the brain.

---

# Neurons

Neurons are like little signal filters. They take incoming signals, convolve them with some sort of a filter, and "spike" when a threshold is reached. This spike sends information downstream to other neurons.

Not only are the neurons spiking, the fluid they're bathed in creates an electrical signal. To make things worse, since your electrode was inserted blind, you could potentially be recording data from multiple neurons at the same time. Before you can model the data, you have to filter the data, then sort the neurons based on their waveform.

---

Finally, you can run regression between the given stimuli, and the processed neural spikes. The simplest model would be a linear model. Certain cells in the brain exhibit mostly linear behavior in response to visual stimuli, so this model works in real data.

```python
import numpy as np
import tables
import scipy.signal as sig
from sklearn import decomposition, mixture, linear_model
from mpl_toolkits.mplot3d import Axes3D
```

```python
h5 = tables.openFile("neurons.hf5")
print h5
plot(h5.root.raw[:1000])
```

```python
b, a = sig.butter(3, 0.1)
lfp = sig.filtfilt(b, a, h5.root.raw[:])
plot(lfp[:1000])
```

```python
data = h5.root.raw[:] - lfp
plot(data[:1000])
```

```python
binary = (data > 5).astype(int)
plot(binary[:1000])
```

```python
leading = np.diff(binary) > 0
times = np.nonzero(leading)[0]
print times[:10]
```

```python
spikes = np.array([data[i-10:i+20] for i in times])
print spikes.shape
lines = plot(spikes[:100].T)
```

```python
pca = decomposition.PCA(3)
pca.fit(spikes)
proj = pca.transform(spikes)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")
ax.scatter(proj[:,0], proj[:,1], proj[:,2])
```

```python
gmm = mixture.GMM(3).fit(proj)
labels = gmm.predict(proj)
print labels[:10]
```

```python
psth = np.zeros((len(data) / 30., len(np.unique(labels))))
for i in np.unique(labels):
	psth[(times[labels == i] / 30).astype(int), i] = 1
plot(psth[:100])
```

```python
stim = h5.root.stim[:]
model = linear_model.ElasticNet(alpha=0.002, l1_ratio=.9)
model.fit(stim, psth)
```

```python
plot(model.coef_.T)
```
