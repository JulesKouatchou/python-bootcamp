```python
# Load the data
data = np.loadtxt("data/stations.txt", dtype=[("station","S4"), 
    ("lat",np.float32), ("lon",np.float32), ("elev",np.float32)])

fig = plt.figure(figsize=(8,6))
fig.suptitle("Seismic Stations", fontsize=22)
ax1 = fig.add_subplot(1,1,1)
sc = ax1.scatter(data["lon"], data["lat"], 
                 s=50*(data["elev"]+1), c=data["elev"], cmap=cm.jet)

# Add text for each station"s 4-letter code
for i in range(data["lon"].size):
    ax1.annotate(data["station"][i], xy=(data["lon"][i], data["lat"][i]), 
        xytext=(data["lon"][i]+(data["elev"][i]/20)**2.2+0.03, data["lat"][i]+(data["elev"][i]/20)**2.2+0.03),
        fontsize=12)
        # Stagger the text offsets as a function the marker size, which is elev

# Add color bar and set its label
cbar = plt.colorbar(sc)
cbar.set_label("Elevation [km]")

ax1.set_xlabel("Longitude [deg]")
ax1.set_ylabel("Latitude [deg]")
```

```python
import sys
class StationPicker(object):
    def __init__(self, figure, stations, eps=0.10, axis=None):
        self.figure = figure
        self.stations = stations
        self.cid = figure.canvas.mpl_connect('button_press_event', self)
        if axis is None:
            axis = figure.axes[0]
        self.axis = axis
        self.eps = eps

    def __call__(self, event):
        #print 'click', event  # dbg
        if event.inaxes != self.axis:
            return
        self.figure.canvas.draw()
        # Compute the distance from the click to all stations
        lats = self.stations['lat']
        longs = self.stations['lon']
        click_long, click_lat = event.xdata, event.ydata
        lat_d = lats - click_lat
        lon_d = longs - click_long
        dist = np.sqrt(lat_d**2 + lon_d**2)
        nearest_i = dist.argmin()
        near_dist = dist[nearest_i]
        nearest = self.stations[nearest_i]
        #print 'Nearest distance:', near_dist  # dbg
        if near_dist < self.eps:
            print "HIT! You clicked on", nearest['station']
        else:
            print "No hit, nearest is:", nearest['station']
            print "It is at:", nearest['lat'], nearest['lon']
            print "Distance to it:", near_dist
        sys.stdout.flush()
```

```python
fig = plt.figure(figsize=(8,6))
fig.suptitle("Seismic Stations", fontsize=22)
ax1 = fig.add_subplot(1,1,1)
sc = ax1.scatter(data["lon"], data["lat"], 
                 s=50*(data["elev"]+1), c=data["elev"], cmap=cm.jet)
cbar = plt.colorbar(sc)
cbar.set_label("Elevation [km]")
ax1.set_xlabel("Longitude [deg]")
ax1.set_ylabel("Latitude [deg]")
# We can now make a picker with that binds the figure and the data
StationPicker(fig, data)
```

```python
img = plt.imread("data/dessert.png")

red, green, blue = [ img[:,:,i] for i in range(3) ]

f, axes = plt.subplots(1, 4, sharex=True, sharey=True, figsize=(12,6))

axes[0].imshow(img, origin="upper")
axes[1].imshow(red, origin="upper", cmap=cm.Reds_r)
axes[2].imshow(green, origin="upper", cmap=cm.Greens_r)
axes[3].imshow(blue, origin="upper", cmap=cm.Blues_r)

# Turn off tick labels
for ax in axes:
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_aspect('auto')
```

```python
f2, axes2 = plt.subplots(4, 1, sharex=True, sharey=True, figsize=(12,10))
lumi = img[:,:,:3].mean(axis=2)
axes2[0].hist(lumi.flatten(), 256, range=[0,1], facecolor='k', edgecolor='k')
axes2[1].hist(red.flatten(), 256, range=[0,1], facecolor='r', edgecolor='r')
axes2[2].hist(green.flatten(), 256, range=[0,1], facecolor='g', edgecolor='g')
axes2[3].hist(blue.flatten(), 256, range=[0,1], facecolor='b', edgecolor='b');
f2.subplots_adjust(hspace=0);

# We want the x tick labels to be invisible on all but the bottom figure
for ax in axes2[:-1]:
    for label in ax.get_xticklabels():
        label.set_visible(False)

# Don't show the last y label in plots 2-4:
for ax in axes2[1:]:
    ax.get_yticklabels()[-1].set_visible(False)

# Clean up the yaxis ticks a bit
axes2[0].set_yticks([2500, 7500, 12500])
```

```python
lumi2 = .3*red + .59*green + 0.11*blue

fig, axes3 = plt.subplots(1, 3, figsize=(12, 8), sharex=True, sharey=True)

axes3[0].imshow(img, origin="upper")
axes3[0].set_title('Color')
axes3[1].imshow(lumi, origin="upper", cmap=cm.Greys_r)
axes3[1].set_title('Naive gray')
axes3[2].imshow(lumi2, origin="upper", cmap=cm.Greys_r)
axes3[2].set_title('Better gray')

for ax in axes3:
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_aspect('auto')
```
